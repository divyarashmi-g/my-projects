Load and visualise data with building baseline models

```python
# loading the dataset  
import pandas as pd
import numpy as np
```

```python
train_set = pd.read_csv('sign_mnist_train.csv')
test_set = pd.read_csv('sign_mnist_test.csv')

# Display the first few rows of each dataset to verify
train_set.head()
```

```python
test_set.head()
```

```python
import matplotlib.pyplot as plt

def plot_letter(data):
    labels = data['label'].unique()
    num_labels = len(labels)
    
    # Set up the figure and axes
    fig, axes = plt.subplots(4, 6, figsize=(12, 8))
    fig.suptitle('Sample of Each Letter', fontsize=16)
    
    for i, label in enumerate(sorted(labels)):
        row = i // 6
        col = i % 6
        ax = axes[row, col]
        
        # Extract the first sample of the current label
        sample = data.loc[data['label'] == label].iloc[0, 1:].values
        sample_image = sample.reshape(32, 32)
        
        # Plot the image and set the title
        ax.imshow(sample_image, cmap='gray')
        ax.set_title(chr(label + 65), fontsize=12)
        ax.axis('off')
    
    # Adjust layout
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Calling the alternative function to plot samples
plot_letter(train_set)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Setting random seeds for reproducibility
import tensorflow as tf

import random
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)
```

```python
num_classes = train_set['label'].max() + 1  

# Preparing the data for the model
X_train = train_set.iloc[:, 1:].values.reshape(-1, 32, 32, 1)  # Adjusting reshape for the correct image size
y_train = to_categorical(train_set['label'].values, num_classes=num_classes)

X_test = test_set.iloc[:, 1:].values.reshape(-1, 32, 32, 1)
y_test = to_categorical(test_set['label'].values, num_classes=num_classes)

# Normalising the data
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# Building a simple baseline model
baseline_model = Sequential([
    Flatten(input_shape=(32, 32, 1)),  
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(num_classes, activation='softmax')  
])

# Compiling the model
baseline_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model
history = baseline_model.fit(X_train, y_train, validation_split=0.1, epochs=10, batch_size=32, verbose=2)

# Evaluating the model on the test data
test_loss, test_accuracy = baseline_model.evaluate(X_test, y_test, verbose=2)
print(f'Test Accuracy: {test_accuracy:.2f}')


```

```python
baseline_model.summary()
```

```python
cnn_model = Sequential([
    Conv2D(16, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(32, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])

# Compiling the model
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model
history_cnn = cnn_model.fit(X_train, y_train, validation_split=0.1, epochs=10, batch_size=32, verbose=2)

# Evaluating the model on the test data
test_loss_cnn, test_accuracy_cnn = cnn_model.evaluate(X_test, y_test, verbose=2)
print(f'Test Accuracy (CNN): {test_accuracy_cnn:.2f}')
```

```python
import matplotlib.pyplot as plt

def plot_model_comparison(history, history_cnn, model1_name="Model 1", model2_name="Model 2"):
    # Plot accuracy
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label=f'{model1_name} Training Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{model1_name} Validation Accuracy')
    plt.plot(history_cnn.history['accuracy'], label=f'{model2_name} Training Accuracy')
    plt.plot(history_cnn.history['val_accuracy'], label=f'{model2_name} Validation Accuracy')
    plt.title('Model Accuracy Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label=f'{model1_name} Training Loss')
    plt.plot(history.history['val_loss'], label=f'{model1_name} Validation Loss')
    plt.plot(history_cnn.history['loss'], label=f'{model2_name} Training Loss')
    plt.plot(history_cnn.history['val_loss'], label=f'{model2_name} Validation Loss')
    plt.title('Model Loss Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

# Assuming history1 and history2 are the history objects of your two models
plot_model_comparison(history, history_cnn, model1_name="Baseline Model", model2_name="Updated CNN Model")
```

```python
from keras.regularizers import l2
from keras.optimizers import Adam

# Defining the CNN v2 model with L2 regularisation
cnn_model_v2_l2 = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 1), kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(num_classes, activation='softmax', kernel_regularizer=l2(0.001))
])

# Compiling the L2 regularized CNN v2 model
cnn_model_v2_l2.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Training the CNN v2 model with L2 regularisation
history_cnn_v2_l2 = cnn_model_v2_l2.fit(X_train, y_train, validation_split=0.1, epochs=20, batch_size=32, verbose=2)

# Evaluating the model on the test data
test_loss_cnn_v2_l2, test_accuracy_cnn_v2_l2 = cnn_model_v2_l2.evaluate(X_test, y_test, verbose=2)
print(f'Final Test Accuracy (CNN v2 with epoch set to 20 and L2 Regularization): {test_accuracy_cnn_v2_l2:.2f}')
```

```python
# Ensure that your model is properly defined and trained before running this code
from sklearn.metrics import accuracy_score

# Check if the model is defined correctly
if isinstance(updated_cnn_model, tf.keras.Model):
    # Predicting the classes for the test set
    y_pred = updated_cnn_model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    # Initialize a dictionary to store accuracies for each alphabet
    alphabet_accuracy = {}

    # Calculate accuracy for each alphabet
    for label in range(num_classes):
        # Identify the indices of the test samples for the current alphabet
        indices = np.where(y_true_classes == label)[0]

        # Calculate the accuracy for the current alphabet
        accuracy = accuracy_score(y_true_classes[indices], y_pred_classes[indices])
        alphabet_accuracy[chr(label + 65)] = accuracy

    # Find the alphabet with the highest and lowest accuracy
    highest_accuracy_alphabet = max(alphabet_accuracy, key=alphabet_accuracy.get)
    lowest_accuracy_alphabet = min(alphabet_accuracy, key=alphabet_accuracy.get)

    # Print the accuracy for each alphabet
    for alphabet, accuracy in alphabet_accuracy.items():
        print(f'Accuracy for {alphabet}: {accuracy:.2f}')

    # Print the highest and lowest accuracy
    print(f'Highest Accuracy: {highest_accuracy_alphabet} with {alphabet_accuracy[highest_accuracy_alphabet]:.2f}')
    print(f'Lowest Accuracy: {lowest_accuracy_alphabet} with {alphabet_accuracy[lowest_accuracy_alphabet]:.2f}')
else:
    print("Error: 'updated_cnn_model' is not recognized as a trained Keras model.")
```
